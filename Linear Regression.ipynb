```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
temp = [18.2, 29.5, 35.9, 50.1, 58.6, 69.9, 82.1, 85.5, 73.5, 55.2, 40.3, 30.1]
daylight = [7.6, 9.2, 12.2, 13.4, 13.9, 14.2, 13.5, 11.8, 10.9, 9.9, 8.4, 7.8]
```

```python
plt.scatter(daylight,temp)
plt.xlabel('Hours of Daylight')
plt.ylabel('Average Monthly Temperature')
plt.show()
```

```python
linregress(daylight,temp)
```

```python
# rvalue = coefficient correlation
```

```python
# pvalue for hypothesis testing
```

```python
# stderr = standard error of the model; the higher the standard error, the less the model describes the data sufficiently
```

```python
# Save the results as a turple; make sure it's the same size as the linear regression
(slope, intercept, rval, pval, stderr) = linregress(daylight,temp)
```

```python
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
# Use list comprehension to get regress_values (average monthly temp given the hours of daylight)
regress_values = [(day * slope + intercept) for day in daylight]
```

```python
plt.scatter(daylight, temp)
plt.plot(daylight, regress_values)
plt.xlabel('Hours of Daylight')
plt.ylabel('Average Monthly Temperature')
plt.show()
```
